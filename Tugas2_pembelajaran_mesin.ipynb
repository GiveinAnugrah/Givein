```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = '/content/Car details v3.csv'
df = pd.read_csv(dataset)
df
```

```python
from matplotlib import pyplot as plt
df['year'].plot(kind='hist', bins=20, title='year')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

frekuensi dari sekitar tahun 1995 sampai sekitar tahun 2005 naik mencapai 250 lalu turun di antara tahun 2005 dan 2010 menjadi dibawah 250, dan dari tahun 2010 sampai tahun 2015 mulai meningkat dari 500 ke 1750 dan mulai menurun di 2020 dengan frekuensi sekitar 500 keatas

```python
from matplotlib import pyplot as plt
import seaborn as sns
df.groupby('owner').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

dari data diatas menunjukkan bahwa jumlah data owner yang paling banyak adalah first owner yaitu mencapai 5000 lebih, dan jumlah data owner yang paling sedikit adalah test drive car sebanyak 0.
